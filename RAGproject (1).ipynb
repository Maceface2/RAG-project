```python
from openai import OpenAI
import numpy as np
from itertools import islice
```

```python
#creates the query and its embedding
client = OpenAI(api_key = "" )
query = "what company does messi endorse"
vectorDB={}
query_embedding = client.embeddings.create(
        input=query,
        model="text-embedding-3-small"
    ).data[0].embedding
print(query_embedding)



```

```python
#parses the txt file and then creates an embedding for each chunk
def embeddingfunction(chuck):
  resource_responses = client.embeddings.create(
            input=chunk,
            model="text-embedding-3-small"
  ).data
  return resource_responses[0].embedding
array=[]
with open("messi.txt", 'r', encoding='utf-8') as file:
    while True:
      chunk = file.read(100)
      if not chunk:
        break
      array+=[chunk]
for item in array:
  vectorDB[item]=embeddingfunction(item)
print(vectorDB)

```

```python
#creates creates the cosine simularity function and makes the vectordb into np arrays

def cosine_simularity(A, B):
    return np.dot(A,B) / ( np.linalg.norm(A) * np.linalg.norm(B) )
for key in vectorDB:
  vectorDB[key]=np.array(vectorDB[key])
similarvectors = vectorDB
print(similarvectors)
```

```python
#you can alter the query here without have to rerun the vectorDB maker
query = "what company does messi endorse"
query_embedding = client.embeddings.create(
        input=query,
        model="text-embedding-3-small"
    ).data[0].embedding
print(query_embedding)
```

```python
#gets a cosine similarity to the query embedding for each chunk and selects the top 5 most relevant chunsk
similars={}
for key in similarvectors.keys():
  similars[key]=cosine_simularity(query_embedding,similarvectors[key])
sorted_similars = dict(sorted(similars.items(), key=lambda item: item[1], reverse=True))
first_5_elements = dict(islice(sorted_similars.items(), 5))
print(first_5_elements)
relevantsentence=""
for key in first_5_elements.keys():
  relevantsentence+=key
```

```python
#inputs those chucks and the query into the llm to create a final output for the user grabbing the needed information that was preselected from the vector database
response = client.chat.completions.create(
    model="gpt-3.5-turbo",  # Use your preferred model
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": query + ":" +relevantsentence},
    ],
)

# Print the response
print(response.choices[0].message.content)
```
